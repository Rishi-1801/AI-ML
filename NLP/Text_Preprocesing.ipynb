```python
import pandas as pd
```

```python
credits=pd.read_csv(r'Datasets\tmdb_5000_credits.csv')
movies=pd.read_csv(r'Datasets\tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['original_title','overview','genres']]
```

```python
movies.head()
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
```

Text Preprocessing

1.lower_casing

```python
movies['overview']=movies['overview'].str.lower()
```

2.Remove Punction Marks

```python
import string
exclude=string.punctuation
```

```python
# Not to do because takes more time
def remove_punc1(text):
    for ch in exclude:
        text=text.replace(ch,'')
    return text
```

```python
# Use this 
def remove_punc2(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
movies['overview']=movies['overview'].apply(remove_punc2)
```

4.Removal of Stop words

```python
import nltk
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))

def remove_stopwords(text):
     return " ".join(
        word for word in text.split()
        if word.lower() not in stop_words
    )
    
```

```python
movies['overview'] = movies['overview'].apply(remove_stopwords)
```

5.Handling Emojis

```python
# not using for this dataset
# for removing emojis
import re
def remove_emoji(text):
    emoji_pattern=re.compile("[" \
    u"\U0001F600-\U0001F64F"
    u"\U0001F300-\U0001F5FF"
    u"\U0001F680-\U0001F6FF"
    u"\U0001F1E0-\U0001F1FF"
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)
```

7.Stemming(making the inflected words to root form)

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

movies['tokens'] = movies['tokens'].apply(
    lambda tokens: [ps.stem(word) for word in tokens]
)
```

Lemmatisation(use this or stemming)

```python
#by spacy
sent3='Running run jogging jog'
doc3=nlp(sent3)
lemmas = [token.lemma_ for token in doc3]
print(lemmas)
```

```python
#by NLTK
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ["players", "playing", "studies", "cars"]
lemmas = [lemmatizer.lemmatize(word) for word in words]
print(lemmas)
```
