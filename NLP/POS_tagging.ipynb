```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(u"I will google about Facebook")
```

```python
doc.text
```

```python
doc[0]
```

```python
doc[1]
```

course grained POS

```python
doc[0].pos_
```

Fine grained POS

```python
doc[0].tag_
```

```python
spacy.explain('PRP')
```

```python
for word in doc:
    print(word.text,'---->',word.pos_,word.tag_,':',spacy.explain(word.tag_))
```

Example of word sense disambuiguition

```python
doc2=nlp(u"I left the room")
for word in doc2:
    print(word.text,'---->',word.pos_,word.tag_,':',spacy.explain(word.tag_))
```

```python
doc3=nlp(u"to the left of the room")
for word in doc3:
    print(word.text,'---->',word.pos_,word.tag_,':',spacy.explain(word.tag_))
```

```python
doc4=nlp(u"I read books on history")
for word in doc4:
    print(word.text,'---->',word.pos_,word.tag_,':',spacy.explain(word.tag_))
```

```python
doc5=nlp(u"I have read book on history")
for word in doc5:
    print(word.text,'---->',word.pos_,word.tag_,':',spacy.explain(word.tag_))
```
