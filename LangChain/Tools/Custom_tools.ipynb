Method 1 - Using tool decorator

```python
from langchain_core.tools import tool
```

```python
# Step 1 : Cerate the function

def multiply(a,b):
    """Multiply two numbers"""     # u need to write this dog string so that in future ur llm understand its use and use it
    return a*b
```

```python
# Step 2: add type hints
def multiply(a:int ,b:int)-> int:
    """Multiply two numbers"""
    return a*b
```

```python
# Step 3: add tool decorator 
@tool
def multiply(a:int,b:int)->int:
    """Multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({'a':3,'b':4})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
# how llm sees ur tool
print(multiply.args_schema.model_json_schema())
```

Method 2- Using the Structured Tool

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):
    a: int =Field(required=True,description='The first number to add')
    b: int =Field(required=True,description='The second number to add')
```

```python
def multiply_fun(a:int,b:int)->int:
    return a*b
```

```python
multiply_tool=StructuredTool.from_function(
    func=multiply_fun,
    name='multiply',
    description='Multiply two numbers',
    args_schema=MultiplyInput
)
```

```python
result=multiply_tool.invoke({"a":3,'b':4})
print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

Method 3- Using Base Tool Class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pydantic 
class MultiplyInput(BaseModel):
    a: int =Field(required=True,description='The first number to add')
    b: int =Field(required=True,description='The second number to add')
```

```python
class MultiplyTool(BaseTool):
    name:str='multiply'
    description:str='Multiply two numbers'
    args_schema:Type[BaseModel]=MultiplyInput

    def _run(self,a:int,b:int)->int:
        return a*b;
```

```python
multiply_tool=MultiplyTool()
result=multiply_tool.invoke({"a":3,'b':4})
print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```
