```python
import ollama
```

Generation

```python
# this prints after overall response is generated
result=ollama.generate(model='llama3.2:1b',prompt='Why does moon shine')
result.response
```

```python
#Stream Response(meaning it will chunks and untill end of response rather than than overall response at once)
response=ollama.generate(model='llama3.2:1b',prompt='Why does moon shine',stream=True)
for i in response:
    print(i.response,end='')
```

```python
# giving the multimodal input to model (Image)
import base64
image_path=r"C:\Users\Rishikesh Reddy\OneDrive\Pictures\Screenshots\Screenshot 2026-05-25 214945.png"
with open(image_path,'rb') as f:   # opening in read mode
    image_bytes=f.read()           # reading the image
image_64=base64.b64encode(image_bytes).decode('utf-8')   # encoding the image to BASE 64 bits
response=ollama.generate(model='gemma3:4b',images=[image_64],prompt='Guess the animal present in the image and its breed?')
print(response.response)
```

```python
# Passing  the multiple images to the model
images_paths=[r"C:\Users\Rishikesh Reddy\OneDrive\Pictures\Screenshots\Screenshot 2026-05-25 214945.png",r"C:\Users\Rishikesh Reddy\OneDrive\Pictures\Screenshots\Screenshot 2026-05-25 194232.png"]
images_base64=[]
for i in images_paths:
    with open(i,'rb') as f:
        image_bytes=f.read()
        images_base64.append(base64.b64encode(image_bytes).decode('utf-8'))

response=ollama.generate(model='gemma3:4b',images=images_base64,prompt='Guess the animals present in the image')
print(response.response)
```

```python
#System Instruction
response=ollama.generate(model='llama3.2:1b',prompt='what is 3 multipled by 10',system='you are an helpful assistant')
print(response.response)
```

```python
#Options parameter(changing parameters)
response=ollama.generate(model='llama3.2:1b',prompt='why is sun appers to be red explain breifly',
                         options={
                             'temperature':0.3,
                             'top_k':45,
                             'top_p':0.7
                         })
print(response.response)
```
