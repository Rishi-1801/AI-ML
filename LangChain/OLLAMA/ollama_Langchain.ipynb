```python
# for chat conversation
from langchain_ollama import ChatOllama

model=ChatOllama(
    model='llama3.2:1b',
    temperature=0.2
)
response=model.invoke('Explain about Black Holes in 5 line Summary')
response.content
```

```python
# For Non Chat Execution
from langchain_ollama import OllamaLLM
model=OllamaLLM(model='llama3.2:1b')
response=model.invoke('The Capital of France')
response
```
