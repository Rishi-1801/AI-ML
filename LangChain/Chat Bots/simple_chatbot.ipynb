```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import HumanMessage,AIMessage
```

```python
model=ChatGroq(model='llama-3.3-70b-versatile')
```

```python
store={}
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]
with_message_history=RunnableWithMessageHistory(model,get_session_history)
```

```python
config={'configurable':{'session_id':'chat1'}}
```

```python
with_message_history.invoke(
    [HumanMessage(content='Hi My name is Rishi and I am a Student')],
    config=config
)
```

```python
with_message_history.invoke(
    [HumanMessage(content="Whta's my name?")],
    config=config
)
```

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
prompt=ChatPromptTemplate.from_messages(
    [('system','You are an helpful Assisstant'),
     MessagesPlaceholder(variable_name='message'),
     ('human','{query}')]
)
chain=prompt|model

# Here system message is for instruction or role
# finally when we invoke===> system,input is created 
# from get_message_history even history is added to the prompt in place of messagePLaceholder
# so prompt= system + input + chat history
# as it is in chain same prompt is passed to llm to get response
```

```python
with_message_history=RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key='query',   # -------> the input query by user
    history_messages_key='message')# -------> the history given by get_session_history
config3={'configurable':{'session_id':'chat3'}}
response=with_message_history.invoke({'query':'My name is Rishi ans I am an AI Enthusiast'},
                                     config=config3)
```

### Managing the Converstion History
One important concept to understand when building chatbots is how to manage conversation history.If left unmanaged, the list of messages will grow unbounded and potentially overflow the context window of the LLM.Therfore, its important to add a step that limits the size of messages you are passing in

```python
# Use the trimmer to trim the whole history and keep recent conversation
from langchain_core.messages import trim_messages
from langchain_core.runnables import RunnablePassthrough
trimmer=trim_messages(
    max_tokens=1000,   # ---> only upto 1000 tokens history
    strategy='last',   # ----> trims starting convo and keeps last
    token_counter=model,  # --> for counting tokens we use model
    include_system=True,   # --> to include always system message
    start_on='human',      #--> if trimmed, after trimming begin from human message
    allow_partial=False    # --> suppose a reply has 5K tokens response which exceeds the maxtokens so there is point that it loads partial msg so it to not happen we made it False
)
```

Final Chain

```python
#trimmed_chain=RunnablePassthrough.assign(
#   message=lambda x:trimmer.invoke(get_session_history(x['configurable']['session_id']).messages)
#)
 
# as th above code gives the error because we dont have x[configurable][session_id] and just have the x['query']
# so to overcome this we manually create the function to add the chathistory database and retrieve it

final_chain = (
    RunnablePassthrough.assign(
        message=lambda x: trimmer.invoke(
            get_session_history(x["session_id"]).messages
        )
    )
    | prompt
    | model
)
```

```python
#with_message_history=RunnableWithMessageHistory(
#   final_chain,
#    get_session_history,
#    input_messages_key='query',
#    history_messages_key='message',
#    )

# Instead of the above code

def chat(session_id, query):
    
    history = get_session_history(session_id)

    response = final_chain.invoke({
        "session_id": session_id,
        "query": query
    })

    history.add_messages([
    HumanMessage(content=query),
    AIMessage(content=response.content)
    ])

    return response.content

```

2nd Method using the RunnableWithMessageHistory

```python
prompt2=ChatPromptTemplate.from_messages(
    [('system','You are an helpful Assisstant'),
     MessagesPlaceholder(variable_name='message'),
    ]
)
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough
chain2=(RunnablePassthrough.assign(message=itemgetter('message')|trimmer)|prompt2|model)
  
```

```python
with_message_history=RunnableWithMessageHistory(
    chain2,
    get_session_history,
    input_messages_key='message'
)
```

```python
response=with_message_history.invoke({
    'message':[HumanMessage(content='Whats my name?')]
},
config=config3)
```
