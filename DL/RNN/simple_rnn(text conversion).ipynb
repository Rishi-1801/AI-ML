Working on docs(Integer encode)

```python
import numpy as np
docs=['go india',
      'india india',
      'hip hip hurray',
      'jeetega bhai jeetega india jeetega',
      'bharat mata ki jai',
      'kohli kohli',
      'sachin sachin',
      'dhoni dhoni',
      'modi ji ki jai',
      'inquilab zindabad']
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()  # if the text comes that is out of traing then it replaces it with "nothing"(oov_token)
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count # no of sentences in document
```

```python
sequences=tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequences,padding='post')
```

```python
sequences
```

```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
model1=Sequential()
model1.add(Embedding(17,output_dim=2,input_length=5))

model1.summary()
```

```python
pred=model1.predict(sequences)      # converts each word to 2 dim vector
print(pred)
```

Working on IMDB Dataset

```python
from keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
print(X_train.shape,"      ",X_test.shape)
```

Padding each sentence

```python
X_train=pad_sequences(X_train,padding='post',maxlen=100)  # input trimed to 100
X_test=pad_sequences(X_test,padding='post',maxlen=100)
```

```python
X_train.shape
```

Building Simple RNN (integer encoded)

```python
model=Sequential()

model.add(SimpleRNN(32,input_shape=(100,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

Building Simple RNN (Embedding)

```python
model2=Sequential()
model2.add(Embedding(10000,2))  # considering vocabulary as 10K
model2.add(SimpleRNN(32,return_sequences=False))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model2.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
