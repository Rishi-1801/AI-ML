```python
import tensorflow as tf
from keras.datasets import imdb
from keras.utils import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding,Bidirectional,SimpleRNN,LSTM,GRU,Dense
```

```python
num_words=10000
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=num_words)

# Pad Sequences
maxlen=100
X_train=pad_sequences(X_train,maxlen=maxlen,padding='post',truncating='post')
X_test=pad_sequences(X_test,maxlen=maxlen,padding='post',truncating='post')

# Buils the Bidirectional Rnn model
embedding_dim=32
model=Sequential()
model.add(Embedding(input_dim=num_words,output_dim=embedding_dim,input_length=maxlen))
model.add(SimpleRNN(5))
model.add(Dense(1,activation='sigmoid'))

#compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#model Summary
model.summary()
```

```python
model=Sequential()
model.add(Embedding(input_dim=num_words,output_dim=embedding_dim,input_length=maxlen))
model.add(Bidirectional(SimpleRNN(5)))
model.add(Dense(1,activation='sigmoid'))

#compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#model Summary
model.summary()
```

```python
# BiLSTM

model=Sequential()
model.add(Embedding(input_dim=num_words,output_dim=embedding_dim,input_length=maxlen))
model.add(Bidirectional(LSTM(5)))
model.add(Dense(1,activation='sigmoid'))

#compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#model Summary
model.summary()
```

```python
# Bidirectional GRU
model=Sequential()
model.add(Embedding(input_dim=num_words,output_dim=embedding_dim,input_length=maxlen))
model.add(Bidirectional(GRU(5)))
model.add(Dense(1,activation='sigmoid'))

#compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#model Summary
model.summary()
```
