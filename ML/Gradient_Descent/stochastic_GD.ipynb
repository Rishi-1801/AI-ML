```python
# it is mainly used for big data as it performs no of rows times update in a single epoch
import numpy as np
```

```python
class SGDRegressor:
    def __init__(self,learning_rate,epochs):
        self.intercept_=None
        self.coef_=None
        self.lr=learning_rate
        self.epochs=epochs
    
    def fit(self,X_train,Y_train):
        self.intercept_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):                                         #repated no of rows times
                idx=np.random.randint(0,X_train.shape[0])                       #some random value
                y_hat=self.intercept_ + np.dot(self.coef_,X_train[idx])         # y cap for that random
                intercept_der=-2*(Y_train[idx]-y_hat)                           # intercept_der for that single row
                coef_der=-2*np.dot(Y_train[idx]-y_hat,X_train[idx])             # coef_der for that single row
                self.intercept_=self.intercept_-(self.lr*intercept_der)         #updated intercept
                self.coef_=self.coef_-(self.lr*coef_der)                        #updated coef

    def predict(self,X_test):
        return self.intercept_ + np.dot(self.coef_,X_test)
```
