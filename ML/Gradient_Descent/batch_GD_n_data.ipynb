```python
#Batch Gardient Descent for n dimensional data
import numpy as np
```

```python
class GDRegressor:
    def __init__(self,learning_rate=0.01,epochs=100):
        self.intercept_=None
        self.coef_=None
        self.lr=learning_rate
        self.epochs=epochs

    def fit(self,X_train,Y_train):
        self.intercept_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epochs):
            y_hat=self.intercept_ + np.dot(X_train,self.coef_)
            intercept_der=-2 * np.mean(Y_train-y_hat)                              #for each iteration it calculates the intercept 
            coef_der=-2 * np.dot((Y_train-y_hat),X_train)/X_train.shape[1]         #np.dot handles automatically transpose and positioning
            self.intercept_=self.intercept_ - (self.lr*intercept_der)              # each iteration it calculates the array of n coefficents
            self.coef_=self.coef_ - (self.lr*coef_der)
        
    def predict(self,X_test):
        return self.intercept_ + np.dot(self.coef_,X_test)
```
