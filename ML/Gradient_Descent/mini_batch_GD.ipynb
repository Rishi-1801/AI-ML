```python
import random
import numpy as np
```

```python
class MBGDRegssor:
    def __init__(self,batch_size,learning_rate,epochs):
        self.intercet_=None
        self.coef_=None
        self.batch_size=batch_size
        self.lr=learning_rate
        self.epochs=epochs

    def fit(self,X_train,Y_train):
        self.intercet_=0
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epochs):
            for j in range(X_train.shape[0]/self.batch_size):
                idx=random.sample(range(X_train.shape[0]),self.batch_size)              #it generates batch no of ids
                y_hat=self.intercet_+np.dot(X_train[idx],self.coef_)                    # all other activities are same as batch but in same particular mini batch wise
                intercept_der=-2*np.mean(Y_train[idx]-y_hat)
                coef_der=-2 * np.dot(Y_train[idx]-y_hat,X_train[idx])/X_train.shape[0]
                self.intercet_=self.intercet_ - (self.lr*intercept_der)
                self.coef_=self.coef_-(self.lr*coef_der)

    def predict(self,X_test):
        return self.intercept_ + np.dot(self.coef_,X_test)
```
