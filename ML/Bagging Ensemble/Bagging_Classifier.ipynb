```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['Species']=iris.target
df.head()
```

```python
new_df=df[df['Species']!=0][['sepal width (cm)',"petal length (cm)",'Species']]
```

```python
new_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(new_df["sepal width (cm)"],new_df["petal length (cm)"],c=new_df["Species"],cmap="winter")
```

```python
# Taking only 10 rows for training
new_df=new_df.sample(100)
df_train=new_df.iloc[:60,:].sample(10)
df_val=new_df.iloc[60:80,:].sample(5)
df_test=new_df.iloc[80:,:].sample(5)
```

```python
df_train.head()
```

```python
X_test=df_val.iloc[:,0:2].values
y_test=df_val.iloc[:,-1].values
```

```python
# Data for Tree 1
df_bag=df_train.sample(8,replace=True)
X=df_bag.iloc[:,0:2]
y=df_bag.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
```

```python
dt_bag1=DecisionTreeClassifier()
```

```python
dt_bag1.fit(X,y)
```

```python
plot_tree(dt_bag1)
plt.show()
y_pred=dt_bag1.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
df_bag=df_train.sample(8,replace=True)
X=df_bag.iloc[:,0:2]
y=df_bag.iloc[:,-1]
dt_bag2=DecisionTreeClassifier()
dt_bag2.fit(X,y)
plot_tree(dt_bag1)
plt.show()
y_pred=dt_bag1.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
