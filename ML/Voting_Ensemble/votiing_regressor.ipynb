```python
import numpy as np
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
df.head()
```

```python
X=df.iloc[:,:13]
y=df.iloc[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()
```

```python
estimators=[('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
    scores=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor(estimators)
scores=cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vr=VotingRegressor(estimators,weights=[i,j,k])
            x=cross_val_score(vr,X,y,cv=10,scoring='r2')
            print(f"for i={i},j={j},k={k}",np.round(np.mean(x),2))
```

```python
# using same Algo

dt1=DecisionTreeRegressor(max_depth=1)
dt2=DecisionTreeRegressor(max_depth=3)
dt3=DecisionTreeRegressor(max_depth=5)
dt4=DecisionTreeRegressor(max_depth=7)
dt5=DecisionTreeRegressor(max_depth=9)

estimators=[('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python
for estimator in estimators:
    scores=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr1=VotingRegressor(estimators)
scores=cross_val_score(vr1,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```
